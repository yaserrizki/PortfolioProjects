# BeuatifulSoup and Requests

```python
from bs4 import BeautifulSoup
import requests
```

```python
# assign the web that we want to scrap into a variable
url = 'https://www.scrapethissite.com/pages/forms/'
```

```python
page = requests.get(url) # send a get request to the URL and it's going to return a response object
page
```

```python
page.text # get the HTML in text type
```

```python
soup = BeautifulSoup(page.text, 'html.parser') # make the HTML more readable
soup
```

```python
print(soup.prettify()) # make the soup structured like a real HTML text
```

# Find and Find_all

```python
soup.find('div') # search for the first 'div' tag in our html
```

```python
soup.find_all('div') # find all 'div' tag in our html and put it on the list
```

```python
soup.find_all('div', class_ = 'col-md-12') #  find all 'div' tag in our html which have CSS class 'col-md-12'
```

```python
# let's say we want to scrap the first text paragraph
# we know from inspect element from our url, the paragraph has <p> tag with the class = "lead"
soup.find_all('p', class_ = 'lead')
```

```python
# now we just wanna extract the paragraph text
# we can use .text attribute in the 'find' module
soup.find('p', class_ = 'lead').text
```

```python
# use .strip module to clean the white space in the beginning and the end of the text
soup.find('p', class_ = 'lead').text.strip()
```

```python
# take a table header in the url
soup.find_all('th')
```

```python
header = soup.find_all('th')
len(header)
```

```python
# take a first table header
data_header = soup.find('th')
data_header = data_header.text.strip
data_header()
```

```python
all_data_header = []
for i in range(len(header)):
    all_data_header.append(header[i].text.strip())

print(all_data_header)
type(all_data_header)
```

# Make a Data Frame with Pandas

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = all_data_header)
df
```

```python
# pull the data from the table and put it into a variable
column_data = soup.find_all('tr')
column_data
```

```python
individual_row_data = []
type(individual_row_data)
for row in column_data[1:]:
    row_data = row.find_all('td')
    for data in row_data:
        individual_row_data.append(data.text.strip())
print(individual_row_data)
```

```python
individual_row_data = []
type(individual_row_data)
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data] # it will scan from td-1 to td-n ... (1)

    length = len(df) # the first iteration will return 0 because our dataframe doesn't have any data yet
    df.loc[length] = individual_row_data # the result of the scan in (1) will be put in the 0th index, then the length of df will be increase by 1

df # after the n-th iteration, the df will be completed
```

```python
# copy the code on the above and modify it, so when we run it, it'll not append the same data
if len(df)!=0:
    df = df.drop(index = df.index) # if df is already filled, we will clear the data first so the length of df will become 0
    if len(df)==0: # after clear the data, we run of the code above
        for row in column_data[1:]:
            row_data = row.find_all('td')
            individual_row_data = [data.text.strip() for data in row_data]
            length = len(df)
            df.loc[length] = individual_row_data

df
```
